```python
#How to use any pretrained model available in tf.keras.applications for PREDICTION !!!!!
# Resnet50
```

```python
import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tf.__version__
```

```python
#Load the model
model = tf.keras.applications.ResNet50(weights="imagenet")
```

```python
#Load an Image
imageToPredict = tf.keras.preprocessing.image.load_img('elephant.jpg', target_size=(224, 224))
plt.imshow(imageToPredict)
```

```python
#Step2: Convert image to np array
myImageNPArray = tf.keras.preprocessing.image.img_to_array(imageToPredict)

myImageNPArray.shape
```

```python
# Step3: Make the image compatible for the model by adding n
myCompatibleImage = np.expand_dims(myImageNPArray, axis = 0)
myCompatibleImage.shape
```

```python
#Step4: Prediction
prediction = model.predict(myCompatibleImage)
```

```python
#Step5: Decode Predictions

tf.keras.applications.resnet50.decode_predictions(prediction , top=5)
```

```python
if tf.keras.applications.resnet50.decode_predictions(prediction , top=5)[0][0][2] > 0.5:
  print(tf.keras.applications.resnet50.decode_predictions(prediction , top=5)[0][0][1])
else:
  print("Model dont know !!!")
```

```python
print(tf.keras.applications.resnet50.decode_predictions(prediction , top=5)[0][0][2])
```

```python
print(tf.keras.applications.resnet50.decode_predictions(prediction , top=5)[0][0][1])
```

```python
tf.keras.applications.resnet50.decode_predictions(prediction , top=5)[0][0]
```
